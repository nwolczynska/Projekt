# Determining consumption safety of mushrooms using Machine Learning

## Before we start:
Before any type of code-based data exploration please carefully read the dataset description file
(agaricus-lepiota.names) and visually inspect the data file (agaricus-lepiota.data) itself to
get a basic 'feel' of the kind of dataset you're going to work with.
### Important takeaways from reading the description file:
    - dataset contains exclusively categorical data
    - missing values are limited to the 'stalk-root' column only 
    - missing values are represented by '?'

### Important takeaways from visual inspection of the data file:
    - dataset does not have a header
    - data entries are non-numerical (will require converting to numbers) 
    
## Importing libraries:


```python
#import libraries
import os
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.impute import KNNImputer
```

## Loading the data:

```python
#specify data directory
os.chdir('C:\\Users\\ROG ZEPHYRUS\\Desktop\\NikoPy')

#provide a list of column names according to dataset description file
ColumnNames = ['consumption-safety', 'cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 'spore-print-color', 'population', 'habitat']

#load the dataset into a Pandas DataFrame
m_data = pd.read_csv('agaricus-lepiota.data', sep=',', header=None, names=ColumnNames, na_values=('?'))
```

## Preprocessing and exploratory analisys:

```python
#check for missing values
print('Number of missing values for each column:\n', m_data.isnull().sum())

m_data.describe()
m_data.info()

#change data type to  categorical 
m_data = m_data.astype('category')
m_data.dtypes

#check unique values for each column
for column in m_data.columns:
    print(column, '\n', m_data[column].unique())
    
#since the variable 'veil-type' has only one uniqe value it will not contribute to the predictive power of tested models, so we can drop it altogether
m_data = m_data.drop(['veil-type'],axis=1)

#use label encoder to transform categorical data into ordinal (since all models work with integers)
encoder=LabelEncoder()  
for column in m_data.columns:
    m_data[column] = encoder.fit_transform(m_data[column])
    
#the missing 'stalk-root' values were also encoded as a separate category, find out which integer are they represented with and replace it with NaN
m_data['stalk-root'] = m_data['stalk-root'].replace(4, np.nan)

#plot every variable divided acording to the consumption_safety
for i, col in enumerate(m_data.columns):
    plt.figure(i)
    sns_plot = sns.countplot(x='consumption-safety', hue=col, data=m_data, palette=('bone'))
    sns.move_legend(sns_plot, "upper left", bbox_to_anchor=(1, 1))
    plt.show()
#plot correlation matrix
correlations = m_data.corr()
plt.subplots(figsize=(24, 18))
data_corr = sns.heatmap(correlations, annot=True, linewidths=0.5, cmap='bone')
plt.show(data_corr)
```

### Missing values handling
There are couple of ways we can deal with missing values, let's explore some of them and see, how they affect classification accuracy.The easiest way is to reduce the dataset by excluding affected data entries or variables
This is not ideal because we wold like to prevent data loss as much as possible but we can give it a try since only one variable contains missing values.

```python
#drop the 'stalk-root' column because of the missing values
m_data_reduced = m_data.drop(['stalk-root'],axis=1)
```

Another common practise is to replace missing values with the overall mean of a given variable, the median or the mode. Since our data is exclusively categorical, the mean doesn't really apply here. Let's use the mode then.

```python
#copy the contents of our dataset, store them in new DataFrame and replace the misiing values with the mode of the 'stalk-root' variable
m_data_mode = m_data.copy()
m_data_mode['stalk-root'].fillna(m_data_mode['stalk-root'].mode()[0], inplace=True)
```

We can also use a classifier to predict the missing value of 'stalk-root' of a given mushroom based on other available characteristics. Since we will eventually try to predict if a given mushroom is safe to eat, we should avoid taking 'consumption-safety' into account at this point.

```python
#delete 'consumption-safety' variable from the dataset
m_data_for_imputer = m_data.drop(['consumption-safety'], axis=1)
#use KNN imputer to preict the missing 'stalk-root' values
imputer = KNNImputer(n_neighbors=1)
m_data_imputed = pd.DataFrame(imputer.fit_transform(m_data_for_imputer), columns=m_data_for_imputer.columns, index=m_data_for_imputer.index)
#Bring back 'consumption-safety' column
m_data_imputed.insert(0,'consumption-safety', m_data['consumption-safety'])
```

### Preparing train/test splits for our datasets

```python
#Prepare an arrays with information about consumption safety only
y_reduced = m_data_reduced['consumption-safety'].values
y_mode = m_data_mode['consumption-safety'].values
y_imputed = m_data_imputed['consumption-safety'].values

# and an arrays with the rest of the features
x_reduced = m_data_reduced.drop(['consumption-safety'], axis=1).values
x_mode = m_data_mode.drop(['consumption-safety'], axis=1).values
x_imputed = m_data_imputed.drop(['consumption-safety'], axis=1).values

# Split data into training and test sets.
x_reduced_train, x_reduced_test, y_reduced_train, y_reduced_test = train_test_split(x_reduced,y_reduced,random_state=17,test_size=0.2)
x_mode_train, x_mode_test, y_mode_train, y_mode_test = train_test_split(x_mode,y_mode,random_state=17,test_size=0.2)
x_imputed_train, x_imputed_test, y_imputed_train, y_imputed_test = train_test_split(x_imputed,y_imputed,random_state=17,test_size=0.2)
```

## Classification

```python
#Logistic Regression
lr_reduced = LogisticRegression(solver='liblinear')
lr_mode = LogisticRegression(solver='liblinear')
lr_imputed = LogisticRegression(solver='liblinear')

lr_reduced.fit(x_reduced_train,y_reduced_train)
lr_mode.fit(x_mode_train,y_mode_train)
lr_imputed.fit(x_imputed_train,y_imputed_train)
print('Logistic Regression Classification')
print('Classification Accuracy for reduced dataset: ', round(lr_reduced.score(x_reduced_test,y_reduced_test)*100,2), '%')
print('Classification Accuracy for mode imputation: ', round(lr_mode.score(x_mode_test,y_mode_test)*100,2), '%')
print('Classification Accuracy for KNN imputation: ', round(lr_imputed.score(x_imputed_test,y_imputed_test)*100,2), '%')

print('Confusion matrices for Logistic Regression')
y_pred_lr_reduced = lr_reduced.predict(x_reduced_test)
y_pred_lr_mode = lr_mode.predict(x_mode_test)
y_pred_lr_imputed = lr_imputed.predict(x_imputed_test)

y_true_lr_reduced = y_reduced_test
y_true_lr_mode = y_mode_test
y_true_lr_imputed = y_imputed_test

cm_reduced = confusion_matrix(y_true_lr_reduced, y_pred_lr_reduced)
cm_mode = confusion_matrix(y_true_lr_mode, y_pred_lr_mode)
cm_imputed = confusion_matrix(y_true_lr_imputed, y_pred_lr_imputed)

f, axs = plt.subplots(ncols=3, figsize =(15,10))
sns.heatmap(cm_reduced,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[0])
sns.heatmap(cm_mode,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[1])
sns.heatmap(cm_imputed,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[2])
axs[0].set(title="Reduced Dataset")
axs[1].set(title="Mode Imputation")
axs[2].set(title="KNN Imputation")
f.text(0.5, 0.04, 'Predicted Values', ha='center')
f.text(0.04, 0.5, 'True values', va='center', rotation='vertical')
plt.show()
```

```python
#Naive Bayes
nb_reduced = GaussianNB()
nb_mode = GaussianNB()
nb_imputed = GaussianNB()

nb_reduced.fit(x_reduced_train,y_reduced_train)
nb_mode.fit(x_mode_train,y_mode_train)
nb_imputed.fit(x_imputed_train,y_imputed_train)
print('Naive Bayes Classification')
print('Classification Accuracy for reduced dataset: ', round(nb_reduced.score(x_reduced_test,y_reduced_test)*100,2), '%')
print('Classification Accuracy for mode imputation: ', round(nb_mode.score(x_mode_test,y_mode_test)*100,2), '%')
print('Classification Accuracy for KNN imputation: ', round(nb_imputed.score(x_imputed_test,y_imputed_test)*100,2), '%')

print('Confusion matrices for Naive Bayes')
y_pred_nb_reduced = nb_reduced.predict(x_reduced_test)
y_pred_nb_mode = nb_mode.predict(x_mode_test)
y_pred_nb_imputed = nb_imputed.predict(x_imputed_test)

y_true_nb_reduced = y_reduced_test
y_true_nb_mode = y_mode_test
y_true_nb_imputed = y_imputed_test

cm_reduced = confusion_matrix(y_true_nb_reduced, y_pred_nb_reduced)
cm_mode = confusion_matrix(y_true_nb_mode, y_pred_nb_mode)
cm_imputed = confusion_matrix(y_true_nb_imputed, y_pred_nb_imputed)

f, axs = plt.subplots(ncols=3, figsize =(15,10))
sns.heatmap(cm_reduced,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[0])
sns.heatmap(cm_mode,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[1])
sns.heatmap(cm_imputed,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[2])
axs[0].set(title="Reduced Dataset")
axs[1].set(title="Mode Imputation")
axs[2].set(title="KNN Imputation")
f.text(0.5, 0.04, 'Predicted Values', ha='center')
f.text(0.04, 0.5, 'True values', va='center', rotation='vertical')
plt.show()
```

```python
optimal_K_value_reduced = 0
optimal_K_value_mode = 0
optimal_K_value_imputed = 0
highest_score_reduced = 0
highest_score_mode = 0
highest_score_imputed = 0

for i in range(1,10):
    knn_reduced = KNeighborsClassifier(n_neighbors=i)
    knn_reduced.fit(x_reduced_train,y_reduced_train)
    if knn_reduced.score(x_reduced_test,y_reduced_test) > highest_score_reduced:
        highest_score_reduced = knn_reduced.score(x_reduced_train,y_reduced_train)
        optimal_K_value_reduced = i
        
for i in range(1,10):
    knn_mode = KNeighborsClassifier(n_neighbors=i)
    knn_mode.fit(x_mode_train,y_mode_train)
    if knn_mode.score(x_mode_test,y_mode_test) > highest_score_mode:
        highest_score_mode = knn_mode.score(x_mode_train,y_mode_train)
        optimal_K_value_mode = i     
               
for i in range(1,10):
    knn_imputed = KNeighborsClassifier(n_neighbors=i)
    knn_imputed.fit(x_imputed_train,y_imputed_train)
    if knn_imputed.score(x_imputed_test,y_imputed_test) > highest_score_imputed:
        highest_score_imputed = knn_imputed.score(x_imputed_train,y_imputed_train)
        optimal_K_value_imputed = i
        
print('KNN Classification')        
print('Optimal K Value for reduced dataset:', optimal_K_value_reduced,'\nClassification Accuracy for reduced dataset: ', round(highest_score_reduced*100,2), '%')
print('Optimal K Value for mode imputation:', optimal_K_value_mode,'\nClassification Accuracy: ', round(highest_score_mode*100,2), '%')
print('Optimal K Value KNN imputation:', optimal_K_value_imputed,'\nClassification Accuracy: ', round(highest_score_imputed*100,2), '%')

print('Confusion matrices for KNN')
y_pred_knn_reduced = knn_reduced.predict(x_reduced_test)
y_pred_knn_mode = knn_mode.predict(x_mode_test)
y_pred_knn_imputed = knn_imputed.predict(x_imputed_test)

y_true_knn_reduced = y_reduced_test
y_true_knn_mode = y_mode_test
y_true_knn_imputed = y_imputed_test

cm_reduced = confusion_matrix(y_true_knn_reduced, y_pred_knn_reduced)
cm_mode = confusion_matrix(y_true_knn_mode, y_pred_knn_mode)
cm_imputed = confusion_matrix(y_true_knn_imputed, y_pred_knn_imputed)

f, axs = plt.subplots(ncols=3, figsize =(15,10))
sns.heatmap(cm_reduced,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[0])
sns.heatmap(cm_mode,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[1])
sns.heatmap(cm_imputed,annot = True, cmap = 'bone', linewidths=0.5,linecolor='black',fmt = '.0f',ax=axs[2])
axs[0].set(title="Reduced Dataset")
axs[1].set(title="Mode Imputation")
axs[2].set(title="KNN Imputation")
f.text(0.5, 0.04, 'Predicted Values', ha='center')
f.text(0.04, 0.5, 'True values', va='center', rotation='vertical')
plt.show()
```
